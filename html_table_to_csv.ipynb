```python
COLUMNS = ['Title', 'Salary/USD', 'Equity %', 'Equity/USD','Work Location','Years of Experience','Years at Startup','Stage','Size']
```

```python
with open('job.html', 'r') as f:

    contents = f.read()

    soup = BeautifulSoup(contents, 'html.parser')
```

```python

table = soup.find("table") # Find the "table" tag in the page
rows = table.find_all("tr") # Find all the "tr" tags in the table
cy_data = []
for row in rows:
    cells = row.find_all("td") #  Find all the "td" tags in each row
#     cells = cells[] # Select the correct columns
    cy_data.append([cell.text for cell in cells]) # For each "td" tag, get the text inside it

df = pd.DataFrame(cy_data, columns=COLUMNS)
```

```python
df.to_csv('jobs.csv')
```
